```python
location_distributions = {
    'Digital Wallet': {'Takeaway': 0.517026, 'In-store': 0.482974},
    'Cash': {'In-store': 0.513909, 'Takeaway': 0.486091},
    'Credit Card': {'In-store': 0.504444, 'Takeaway': 0.495556},
}

# Function to fill NaN values based on distribution
def fill_na_location(row):
    if pd.isna(row['Location']):
        payment_method = row['Payment Method']
        if payment_method in location_distributions:
            choices, probabilities = zip(*location_distributions[payment_method].items())
            return np.random.choice(choices, p=probabilities)
    return row['Location']

# Apply function to fill NaN values
cafe['Location'] = cafe.apply(fill_na_location, axis=1)
```
