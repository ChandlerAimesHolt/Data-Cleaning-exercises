```python
employees['Performance level'].unique()
```

```python
employees['Performance level'] = employees['Performance level'].replace('Outstanding', '5')
employees['Performance level'] = employees['Performance level'].replace('Exceeds', '4')
employees['Performance level'] = employees['Performance level'].replace('Meets', '3')
employees['Performance level'] = employees['Performance level'].replace('Needs Improvement', '2')
employees['Performance level'] = employees['Performance level'].replace('Below', '1')
```

```python
employees['Performance level'].unique()
```

```python
employees['Salary'].sort_values()
```

```python
employees['Department'].sort_values().unique()
```

```python
employees['Department'] = employees['Department'].replace('Customer Servic', 'Customer Service')
employees['Department'] = employees['Department'].replace('Financ', 'Finance')
employees['Department'] = employees['Department'].replace('H', 'HR')
employees['Department'] = employees['Department'].replace('I', 'IT')
employees['Department'] = employees['Department'].replace('marketinging', 'Marketing')
employees['Department'] = employees['Department'].replace('Operation', 'Operations')
employees['Department'] = employees['Department'].replace('Sale', 'Sales')
employees['Department'] = employees['Department'].replace('engineeringing', 'Engineering')
```

```python
employees['Department'].sort_values().unique()
```

```python
from datetime import datetime
#checking to make sure there are not any start dates that came after the resignation date
(pd.to_datetime(employees['Date Joined'], format='%Y-%m-%d') > pd.to_datetime(employees['Date Resigned'], format='%Y-%m-%d')).sum()
```

```python
from dateutil import parser

employees['DOB'] = (employees['DOB'].apply(lambda x: parser.parse(x) if pd.notnull(x) else pd.NaT))
```

```python
employees['Gender'].unique()
```

```python
employees['Gender'] = employees['Gender'].replace({'Woman' : 'F',
                             'Man' : 'M',
                             'Girl' : 'F',
                             'Boy' : 'M',
                             'Decline to comment' : np.nan,
                             'Male' : 'M',
                             'Female' : 'F'})
```

```python
employees['Gender'] = employees['Gender'].fillna('No Thanks')
```

```python
employees['Age'].unique()
```

```python
employees['Age'] = employees['Age'].fillna(0)
employees['Age'] = employees['Age'].astype('int')
```

```python
current_year = datetime.today().year

employees.loc[employees['DOB'].isna(), 'DOB'] = employees['Age'].apply(lambda x: datetime(current_year - x, 1, 1))
```

```python
employees.isna().sum()
```

```python
employees[['Age','DOB']]
```

```python
employees['DOB'] = pd.to_datetime(employees['DOB'], errors='coerce')

current_year = datetime.today().year

employees['Correct_Birth_Year'] = current_year - employees['Age']

employees['DOB'] = employees.apply(lambda row: row['DOB'].replace(year=row['Correct_Birth_Year']) if pd.notna(row['DOB']) else row['DOB'], axis=1)

employees.drop(columns=['Correct_Birth_Year'], inplace=True)

employees
```
