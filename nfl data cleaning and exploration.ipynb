```python
players.columns = ['index', 'Rnd', 'Player', 'Year', 'College', 'POS', 'Height (in)',
       'Weight (lbs)', 'Hand Size (in)', 'Arm Length (in)', 'Wonderlic',
       '40 Yard', 'Bench Press', 'Vert Leap (in)', 'Broad Jump (in)',
       'Shuttle', '3Cone', '60Yd Shuttle']
```

```python
players = players.fillna(players.groupby('POS')[['Hand Size (in)', 'Arm Length (in)', 'Wonderlic',
       '40 Yard', 'Bench Press', 'Vert Leap (in)', 'Broad Jump (in)',
       'Shuttle', '3Cone', '60Yd Shuttle']].transform('median'))
```
